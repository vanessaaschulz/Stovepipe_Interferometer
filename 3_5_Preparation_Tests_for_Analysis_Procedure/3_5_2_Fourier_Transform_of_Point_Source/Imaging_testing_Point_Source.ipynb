```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib.image as mpimg
```

```python
def pointSourceFFT(imgSize, ypos, xpos, amp=10.):
    img = np.zeros((imgSize+1, imgSize+1)) #odd sized array such that there is a central pixel
    img[ypos,xpos] = amp #make the central pixel have an intensity of 1
    fftImg = np.fft.fft2(np.fft.fftshift(img)) #compute the Fourier transform of the image
    
    fig, axes = plt.subplots(figsize=(16,8))
    plt.subplot(1,2,1)
    plt.title('Image')
    plt.imshow(img, interpolation='nearest')
    plt.set_cmap('gray')
    plt.colorbar(shrink=0.5)
    
    plt.subplot(1,2,2)
    plt.title('Fourier Transform (phase)')
    plt.imshow(np.angle(fftImg))
    plt.set_cmap('hsv')
    plt.colorbar(shrink=0.5)
    plt.savefig('Pixel.png',dpi=200,bbox_inches='tight')
    #amplitudes are all 1.
    print('FFT Max:', np.max(np.abs(fftImg)), 'Min:', np.min(np.abs(fftImg)))

imgSize = 128
pointSourceFFT(imgSize, int(imgSize/2)+1, int(imgSize/2)+1)
imgSize = 128
#pointSourceFFT(imgSize, int(imgSize/2)+20, int(imgSize/2))
imgSize = 128
#pointSourceFFT(imgSize, int(imgSize/2), int(imgSize/2)+20)
imgSize = 128
#pointSourceFFT(imgSize, int(imgSize/2)+2, int(imgSize/2))
```

```python
def multipleSourcesFFT(imgSize, pos, amp):
    img = np.zeros((imgSize+1, imgSize+1)) #odd sized array such that there is a central pixel
    for p,a in zip(pos, amp):
        img[p[0], p[1]] = a #make the central pixel have an intensity of 1
    
    fftImg = np.fft.fft2(np.fft.fftshift(img)) #compute the Fourier transform of the image
    
    fig, axes = plt.subplots(figsize=(16,8))
    plt.subplot(1,2,1)
    plt.title('Image')
    plt.imshow(img, interpolation='nearest')
    plt.set_cmap('gray')
    plt.colorbar(shrink=0.5)
    
    plt.subplot(1,2,2)
    plt.title('Fourier Transform (phase)')
    plt.imshow(np.angle(fftImg))
    plt.set_cmap('hsv')
    plt.colorbar(shrink=0.5)
    
    #amplitudes are all 1.
    print('FFT Max:', np.max(np.abs(fftImg)), 'Min:', np.min(np.abs(fftImg)))

imgSize = 128

multipleSourcesFFT(imgSize, [[64, 65], [90,65]], [1., 1.])

multipleSourcesFFT(imgSize, [[20, 21], [101,80]], [1., 1.])

multipleSourcesFFT(imgSize, [[64, 20], [64,100]], [1., 1.])
```

```python
multipleSourcesFFT(imgSize, [[64, 65], [90,65]], [1., 0.1])
multipleSourcesFFT(imgSize, [[64, 65], [90,65]], [1., 1.0])
multipleSourcesFFT(imgSize, [[64, 65], [90,65]], [1., 10.])

multipleSourcesFFT(imgSize, [[20, 21], [101,80]], [1., 0.1])
multipleSourcesFFT(imgSize, [[20, 21], [101,80]], [1., 1.])
multipleSourcesFFT(imgSize, [[20, 21], [101,80]], [1., 10.])

multipleSourcesFFT(imgSize, [[64, 20], [64,100]], [1., 0.1])
multipleSourcesFFT(imgSize, [[64, 20], [64,100]], [1., 1.])
multipleSourcesFFT(imgSize, [[64, 20], [64,100]], [1., 10.])
```
